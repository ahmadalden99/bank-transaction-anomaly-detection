#### Sine-Cosine Encoding pada Date Features

Pada tahap feature engineering, kolom tanggal tidak langsung digunakan dalam bentuk mentah. Kolom tanggal seperti `TransactionDate` awalnya berisi informasi waktu transaksi dalam format datetime, misalnya `2023-04-11 16:29:14`. Format seperti ini tidak langsung digunakan sebagai input model karena model machine learning tidak memahami makna waktu secara langsung dari format datetime.

Oleh karena itu, informasi waktu dari `TransactionDate` diubah menjadi beberapa fitur baru yang lebih mudah dipahami oleh model, yaitu:

```text
Hour
DayOfWeek
Month
IsWeekend
```

Fitur `Hour` menunjukkan jam transaksi terjadi, `DayOfWeek` menunjukkan hari transaksi dalam satu minggu, `Month` menunjukkan bulan transaksi, dan `IsWeekend` menunjukkan apakah transaksi terjadi pada akhir pekan.

Namun, fitur seperti `Hour`, `DayOfWeek`, dan `Month` memiliki sifat khusus, yaitu bersifat siklik. Artinya, nilai fitur tersebut membentuk siklus atau putaran.

Contohnya:

```text
Hour:
0, 1, 2, ..., 22, 23, kembali ke 0

DayOfWeek:
Senin, Selasa, ..., Minggu, kembali ke Senin

Month:
Januari, Februari, ..., Desember, kembali ke Januari
```

Karena sifatnya yang berulang, fitur waktu tidak ideal jika digunakan sebagai angka mentah.

#### Masalah Jika Menggunakan Angka Mentah

Jika fitur waktu digunakan sebagai angka biasa, model dapat salah memahami jarak antar nilai.

Contoh pada fitur `Hour`:

```text
Hour 23 dan Hour 0
```

Secara angka biasa:

```text
23 - 0 = 23
```

Model dapat menganggap jam 23 dan jam 0 sebagai dua waktu yang sangat jauh. Padahal secara waktu, jam 23:00 dan 00:00 hanya berbeda satu jam.

Masalah yang sama juga terjadi pada `DayOfWeek` dan `Month`.

Pada `DayOfWeek`, hari Minggu dan Senin sebenarnya berdekatan karena setelah Minggu siklus kembali ke Senin. Namun, jika direpresentasikan sebagai angka 6 dan 0, model dapat menganggap keduanya jauh.

Pada `Month`, bulan Desember dan Januari juga berdekatan karena setelah Desember siklus kembali ke Januari. Namun, jika direpresentasikan sebagai angka 12 dan 1, model dapat menganggap keduanya jauh.

Oleh karena itu, fitur waktu perlu diubah ke bentuk yang dapat mempertahankan sifat sikliknya.

#### Konsep Sine-Cosine Encoding

Sine-cosine encoding digunakan untuk mengubah fitur siklik menjadi representasi koordinat pada lingkaran.

Logikanya, fitur waktu seperti jam, hari, dan bulan tidak berbentuk garis lurus, tetapi berbentuk lingkaran. Setelah mencapai nilai terakhir dalam satu siklus, nilainya kembali lagi ke awal.

Dengan sine-cosine encoding, setiap nilai waktu diubah menjadi dua nilai baru:

```text
sin value
cos value
```

Dua nilai ini merepresentasikan posisi suatu titik pada lingkaran.

Secara sederhana:

```text
sin = posisi vertikal pada lingkaran
cos = posisi horizontal pada lingkaran
```

Gabungan nilai sine dan cosine membuat model dapat memahami posisi waktu dalam satu siklus secara lebih tepat.

#### Rumus Umum

Rumus umum sine-cosine encoding adalah sebagai berikut:

$$
\mathrm{sin_value} = \sin\left(2\pi \times \frac{\mathrm{value}}{\mathrm{period}}\right)
$$

$$
\mathrm{cos_value} = \cos\left(2\pi \times \frac{\mathrm{value}}{\mathrm{period}}\right)
$$

Keterangan:

```text
value  = nilai fitur waktu
period = panjang satu siklus penuh
```

Contoh period:

```text
Hour       = 24
DayOfWeek  = 7
Month      = 12
```

Faktor `2π` digunakan karena satu putaran penuh pada lingkaran sama dengan `2π` radian.

#### Hour Sine-Cosine Encoding

Fitur `Hour` menunjukkan jam transaksi terjadi. Nilainya berada pada rentang 0 sampai 23.

Karena satu hari memiliki 24 jam, maka period untuk fitur `Hour` adalah 24.

Fitur `Hour` diubah menjadi dua fitur baru:

```text
Hour_sin
Hour_cos
```

Rumus yang digunakan adalah:

$$
\mathrm{Hour_sin} = \sin\left(2\pi \times \frac{\mathrm{Hour}}{24}\right)
$$

$$
\mathrm{Hour_cos} = \cos\left(2\pi \times \frac{\mathrm{Hour}}{24}\right)
$$

Contoh interpretasi:

| Hour |    Hour_sin |    Hour_cos | Makna              |
| ---: | ----------: | ----------: | ------------------ |
|    0 |           0 |           1 | awal hari          |
|    6 |           1 |           0 | seperempat hari    |
|   12 |           0 |          -1 | tengah hari        |
|   18 |          -1 |           0 | malam              |
|   23 | mendekati 0 | mendekati 1 | dekat dengan jam 0 |

Nilai negatif pada `Hour_sin` atau `Hour_cos` adalah hal yang normal. Nilai sine dan cosine memang berada pada rentang -1 sampai 1.

Dengan transformasi ini, model dapat memahami bahwa jam 23 dekat dengan jam 0, karena keduanya berada pada posisi yang berdekatan di lingkaran.

#### DayOfWeek Sine-Cosine Encoding

Fitur `DayOfWeek` menunjukkan hari transaksi dalam satu minggu. Pada pandas, nilai `DayOfWeek` biasanya berada pada rentang 0 sampai 6.

```text
0 = Senin
1 = Selasa
2 = Rabu
3 = Kamis
4 = Jumat
5 = Sabtu
6 = Minggu
```

Karena satu minggu memiliki 7 hari, maka period untuk fitur `DayOfWeek` adalah 7.

Fitur `DayOfWeek` diubah menjadi dua fitur baru:

```text
DayOfWeek_sin
DayOfWeek_cos
```

Rumus yang digunakan adalah:

$$
\mathrm{DayOfWeek_sin} = \sin\left(2\pi \times \frac{\mathrm{DayOfWeek}}{7}\right)
$$

$$
\mathrm{DayOfWeek_cos} = \cos\left(2\pi \times \frac{\mathrm{DayOfWeek}}{7}\right)
$$

Transformasi ini digunakan agar model memahami bahwa hari Minggu dan Senin adalah hari yang berdekatan dalam satu siklus mingguan.

Jika `DayOfWeek` digunakan sebagai angka mentah, model dapat menganggap nilai 6 dan 0 sebagai nilai yang jauh. Padahal secara siklus minggu, setelah Minggu akan kembali ke Senin.

Selain sine-cosine encoding, dibuat juga fitur `IsWeekend`.

Fitur `IsWeekend` digunakan untuk menandai apakah transaksi terjadi pada akhir pekan.

Aturannya:

```text
Sabtu atau Minggu = 1
Senin sampai Jumat = 0
```

Fitur ini tetap digunakan karena dapat membantu model membedakan transaksi pada hari kerja dan akhir pekan.

#### Month Sine-Cosine Encoding

Fitur `Month` menunjukkan bulan transaksi. Nilainya berada pada rentang 1 sampai 12.

```text
1  = Januari
2  = Februari
3  = Maret
...
12 = Desember
```

Karena satu tahun memiliki 12 bulan, maka period untuk fitur `Month` adalah 12.

Fitur `Month` diubah menjadi dua fitur baru:

```text
Month_sin
Month_cos
```

Rumus yang digunakan adalah:

$$
\mathrm{Month_sin} = \sin\left(2\pi \times \frac{\mathrm{Month}}{12}\right)
$$

$$
\mathrm{Month_cos} = \cos\left(2\pi \times \frac{\mathrm{Month}}{12}\right)
$$

Transformasi ini digunakan agar model memahami bahwa bulan Desember dekat dengan Januari.

Jika `Month` digunakan sebagai angka mentah, model dapat menganggap bulan 12 dan bulan 1 sebagai nilai yang jauh. Padahal secara siklus tahunan, setelah Desember akan kembali ke Januari.

Dengan sine-cosine encoding, hubungan siklik antar bulan tetap terjaga.

#### Mengapa Harus Menggunakan Sin dan Cos Sekaligus?

Sine-cosine encoding menggunakan dua fitur karena satu fitur saja tidak cukup untuk menunjukkan posisi dalam lingkaran secara lengkap.

Jika hanya menggunakan sine, beberapa nilai waktu yang berbeda dapat memiliki nilai sine yang sama atau mirip. Hal ini dapat membuat model sulit membedakan posisi waktu tersebut.

Dengan menambahkan cosine, posisi waktu dapat direpresentasikan sebagai titik dua dimensi pada lingkaran.

Secara sederhana:

```text
sin saja = informasi belum lengkap
sin + cos = posisi siklus lebih lengkap
```

Oleh karena itu, setiap fitur siklik diubah menjadi dua fitur baru.

Contohnya:

```text
Hour       menjadi Hour_sin dan Hour_cos
DayOfWeek  menjadi DayOfWeek_sin dan DayOfWeek_cos
Month      menjadi Month_sin dan Month_cos
```

#### Mengapa Nilainya Bisa Negatif?

Nilai sine dan cosine memang berada pada rentang:

```text
-1 sampai 1
```

Jadi jika hasil encoding bernilai negatif, hal itu bukan error.

Contoh:

```text
Hour_sin = -0.866
Hour_cos = -0.500
```

Nilai tersebut hanya menunjukkan posisi waktu pada bagian tertentu dari lingkaran.

Hal ini berbeda dengan fitur seperti `TimeSincePreviousTransaction_Hours`. Fitur tersebut seharusnya bernilai positif karena menunjukkan selisih waktu. Jika fitur selisih waktu bernilai negatif, maka ada kemungkinan masalah pada data tanggal.

Namun, untuk fitur sine-cosine seperti `Hour_sin`, `Hour_cos`, `DayOfWeek_sin`, `DayOfWeek_cos`, `Month_sin`, dan `Month_cos`, nilai negatif adalah hal yang normal.

### 2. Amount dan balance features

1. Berapa besar nominal transaksi aslinya?
2. Seberapa besar transaksi itu dibanding saldo akun?
3. Apakah nominal transaksi itu wajar dibanding transaksi lain yang jenisnya sama dan kelompok saldonya mirip?

Amount_To_Balance_Ratio = TransactionAmount / AccountBalance

Amount_To_TypeBalanceGroupAvg_Ratio -->
BalanceGruopGlobal --> Low, Medium, High

````python
import numpy as np
import pandas as pd
from IPython.display import display

if "df_fe" not in globals():
    df_fe = df.copy()


required_columns = [
    "TransactionAmount",
    "AccountBalance",
    "TransactionType",
    "Channel"
]

missing_columns = [col for col in required_columns if col not in df_fe.columns]

if len(missing_columns) > 0:
    raise ValueError(f"Kolom berikut tidak ditemukan: {missing_columns}")

print("Kolom yang digunakan:")
display(df_fe[required_columns].head())


df_fe["Amount_to_Balance_Ratio"] = (
    df_fe["TransactionAmount"] / df_fe["AccountBalance"].replace(0, np.nan)
)


df_fe["BalanceRank_ByType"] = (
    df_fe
    .groupby("TransactionType")["AccountBalance"]
    .rank(method="first", pct=True)
)

df_fe["BalanceGroup_ByType"] = pd.cut(
    df_fe["BalanceRank_ByType"],
    bins=[0, 1/3, 2/3, 1],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)


balance_thresholds = (
    df_fe
    .groupby("TransactionType")["AccountBalance"]
    .quantile([1/3, 2/3])
    .unstack()
)

balance_thresholds.columns = [
    "Low_to_Medium_Boundary",
    "Medium_to_High_Boundary"
]



avg_amount_by_type_channel_balance_group = (
    df_fe
    .groupby(
        ["TransactionType", "Channel", "BalanceGroup_ByType"],
        observed=True
    )["TransactionAmount"]
    .transform("mean")
)


df_fe["Amount_to_TypeChannelBalanceGroupAvg_Ratio"] = (
    df_fe["TransactionAmount"] / avg_amount_by_type_channel_balance_group.replace(0, np.nan)
)



amount_balance_features = [
    "TransactionAmount",
    "AccountBalance",
    "Amount_to_Balance_Ratio",
    "Amount_to_TypeChannelBalanceGroupAvg_Ratio"
]

df_fe[amount_balance_features] = (
    df_fe[amount_balance_features]
    .replace([np.inf, -np.inf], np.nan)
)



print("=" * 70)
print("Jumlah data per TransactionType dan BalanceGroup_ByType")
print("=" * 70)
display(pd.crosstab(df_fe["TransactionType"], df_fe["BalanceGroup_ByType"]))

print("\n" + "=" * 70)
print("Batas saldo Low, Medium, dan High per TransactionType")
print("=" * 70)
display(balance_thresholds)

print("\n" + "=" * 70)
print("Rata-rata TransactionAmount berdasarkan TransactionType + Channel + BalanceGroup_ByType")
print("=" * 70)
display(
    df_fe
    .groupby(
        ["TransactionType", "Channel", "BalanceGroup_ByType"],
        observed=True
    )["TransactionAmount"]
    .mean()
    .reset_index()
    .rename(columns={"TransactionAmount": "Avg_TransactionAmount"})
)

print("\n" + "=" * 70)
print("Statistik fitur amount dan balance")
print("=" * 70)
display(df_fe[amount_balance_features].describe())

print("\n" + "=" * 70)
print("Missing value pada fitur amount dan balance")
print("=" * 70)
display(df_fe[amount_balance_features].isnull().sum())

print("\n" + "=" * 70)
print("Contoh hasil fitur amount dan balance")
print("=" * 70)
display(
    df_fe[
        [
            "TransactionType",
            "Channel",
            "TransactionAmount",
            "AccountBalance",
            "Amount_to_Balance_Ratio",
            "BalanceRank_ByType",
            "BalanceGroup_ByType",
            "Amount_to_TypeChannelBalanceGroupAvg_Ratio"
        ]
    ].head(10)
)
````

**Amount and Balance Feature Engineering**

Pada tahap ini, fitur yang berkaitan dengan nominal transaksi dan saldo akun digunakan untuk memberi konteks terhadap besar kecilnya suatu transaksi. Nominal transaksi tidak cukup hanya dilihat dari nilai aslinya, karena transaksi dengan jumlah yang sama dapat memiliki makna berbeda tergantung pada saldo akun, jenis transaksi, dan channel transaksi yang digunakan.

Sebagai contoh, transaksi sebesar `900` dapat dianggap besar jika saldo akun hanya `1.000`, tetapi dapat dianggap kecil jika saldo akun sebesar `100.000`. Oleh karena itu, dibuat beberapa fitur tambahan agar model tidak hanya membaca nominal transaksi secara mentah, tetapi juga memahami konteks transaksi tersebut.

**Fitur Dasar yang Digunakan**

Fitur dasar yang digunakan pada tahap ini adalah:

```text
TransactionAmount
AccountBalance
TransactionType
Channel
```

`TransactionAmount` menunjukkan nominal transaksi yang dilakukan oleh customer.

`AccountBalance` menunjukkan saldo akun customer pada saat transaksi.

`TransactionType` menunjukkan jenis transaksi, misalnya `Debit` atau `Credit`.

`Channel` menunjukkan jalur transaksi, misalnya `Online`, `ATM`, atau `Branch`.

**Amount to Balance Ratio**

Fitur pertama yang dibuat adalah:

```text
Amount_to_Balance_Ratio
```

Fitur ini digunakan untuk mengukur seberapa besar nominal transaksi dibandingkan dengan saldo akun.

Rumus konsepnya:

```text
Amount_to_Balance_Ratio = TransactionAmount / AccountBalance
```

Contoh pertama:

```text
TransactionAmount = 900
AccountBalance = 1.000
Amount_to_Balance_Ratio = 900 / 1.000
Amount_to_Balance_Ratio = 0,90
```

Artinya, transaksi tersebut menggunakan 90% dari saldo akun.

Contoh kedua:

```text
TransactionAmount = 900
AccountBalance = 100.000
Amount_to_Balance_Ratio = 900 / 100.000
Amount_to_Balance_Ratio = 0,009
```

Artinya, transaksi tersebut hanya menggunakan 0,9% dari saldo akun.

Dari contoh tersebut, dapat dilihat bahwa nominal transaksi yang sama dapat memiliki makna berbeda jika dibandingkan dengan saldo akun. Karena itu, fitur `Amount_to_Balance_Ratio` dibuat agar model dapat memahami proporsi transaksi terhadap saldo.

**Balance Group by Transaction Type**

Fitur berikutnya yang dibuat adalah:

```text
BalanceGroup_ByType
```

Fitur ini digunakan untuk membagi saldo akun menjadi tiga kelompok:

```text
Low
Medium
High
```

Pembagian kelompok saldo tidak dilakukan secara global, tetapi dilakukan di dalam masing-masing `TransactionType`.

Artinya, saldo pada transaksi `Debit` dibandingkan dengan saldo transaksi `Debit` lain. Sementara itu, saldo pada transaksi `Credit` dibandingkan dengan saldo transaksi `Credit` lain.

Alurnya:

```text
Debit
  saldo 33% terbawah = Low
  saldo 33% tengah   = Medium
  saldo 33% teratas  = High

Credit
  saldo 33% terbawah = Low
  saldo 33% tengah   = Medium
  saldo 33% teratas  = High
```

Dengan cara ini, kelompok saldo menjadi lebih adil karena saldo dibandingkan dengan transaksi yang memiliki jenis transaksi sama.

Sebagai contoh, `Low` pada transaksi `Debit` berarti saldo rendah dibandingkan transaksi `Debit` lain. Sedangkan `Low` pada transaksi `Credit` berarti saldo rendah dibandingkan transaksi `Credit` lain.

**Penggunaan Channel sebagai Konteks Pembanding**

Selain `TransactionType`, fitur `Channel` juga digunakan sebagai konteks pembanding. Hal ini karena transaksi melalui `Online`, `ATM`, dan `Branch` dapat memiliki pola nominal yang berbeda.

Sebagai contoh, transaksi dengan nominal besar mungkin masih wajar jika dilakukan melalui `Branch`, tetapi dapat terlihat tidak biasa jika dilakukan melalui `ATM`.

Namun, `Channel` tidak digunakan untuk membentuk kelompok saldo `Low`, `Medium`, dan `High`. Kelompok saldo tetap dibuat berdasarkan `TransactionType`.

Setelah kelompok saldo terbentuk, `Channel` digunakan untuk menghitung rata-rata nominal transaksi pada kelompok yang lebih spesifik.

Kelompok pembanding yang digunakan adalah:

```text
TransactionType + Channel + BalanceGroup_ByType
```

Artinya, suatu transaksi dibandingkan dengan transaksi lain yang memiliki:

```text
jenis transaksi yang sama
channel transaksi yang sama
kelompok saldo yang sama
```

**Amount to Type Channel Balance Group Average Ratio**

Fitur utama berikutnya adalah:

```text
Amount_to_TypeChannelBalanceGroupAvg_Ratio
```

Fitur ini digunakan untuk membandingkan nominal transaksi dengan rata-rata nominal transaksi pada kelompok yang sama.

Kelompok yang sama ditentukan berdasarkan:

```text
TransactionType
Channel
BalanceGroup_ByType
```

Rumus konsepnya:

```text
Ratio = TransactionAmount / rata-rata TransactionAmount pada kelompok yang sama
```

Contoh:

```text
TransactionType = Debit
Channel = ATM
BalanceGroup_ByType = Low
TransactionAmount = 900
```

Misalnya rata-rata `TransactionAmount` untuk kelompok `Debit + ATM + Low` adalah `150`.

Maka:

```text
Ratio = 900 / 150
Ratio = 6
```

Artinya, transaksi tersebut 6 kali lebih besar dibandingkan rata-rata transaksi `Debit` melalui `ATM` dari akun dengan kelompok saldo rendah.

Fitur ini penting karena model tidak hanya melihat nominal transaksi secara mentah, tetapi juga membandingkannya dengan transaksi lain yang memiliki konteks serupa.

## Modeling / Anomaly Detection

Karena dataset kita tidak punya label fraud asli, jadi kita tidak bikin supervised classification dulu. Kita pakai unsupervised anomaly detection

### Setup modeling dan drop zero variance

````python
import numpy as np
import pandas as pd
from IPython.display import display

from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.covariance import EllipticEnvelope

if "X_model" not in globals():
    raise ValueError("X_model belum ada. Jalankan preprocessing dulu.")

if "df_fe" not in globals():
    raise ValueError("df_fe belum ada. Jalankan feature engineering dulu.")

zero_variance_features = X_model.columns[X_model.nunique() <= 1].tolist()

X_model_clean = X_model.drop(columns=zero_variance_features).copy()

X_model_clean = X_model_clean.astype(float)

contamination_rate = 0.05
random_state = 42

print("Fitur zero variance yang dihapus:")
print(zero_variance_features)

print("\nShape X_model sebelum drop:")
print(X_model.shape)

print("\nShape X_model_clean setelah drop:")
print(X_model_clean.shape)

print("\nContamination rate:")
print(contamination_rate)
````

### Siapkan dataframe hasil modeling

````python
result_columns = [
    "TransactionID",
    "AccountID",
    "TransactionAmount",
    "AccountBalance",
    "TransactionType",
    "Channel",
    "Location",
    "CustomerAge",
    "CustomerOccupation",
    "TransactionDuration",
    "LoginAttempts"
]

existing_result_columns = [
    col for col in result_columns
    if col in df_fe.columns
]

df_result = df_fe[existing_result_columns].copy()

print("Shape df_result:")
print(df_result.shape)

print("\nKolom df_result:")
display(pd.DataFrame({"Column": df_result.columns}))

print("\nContoh df_result:")
display(df_result.head())
````

### Isolation Forest

````python
if_model = IsolationForest(
    n_estimators=300,
    contamination=contamination_rate,
    random_state=random_state
)

if_model.fit(X_model_clean)

if_pred = if_model.predict(X_model_clean)
if_score = -if_model.decision_function(X_model_clean)

df_result["IF_Anomaly"] = np.where(if_pred == -1, 1, 0)
df_result["IF_Score"] = if_score

print("Jumlah hasil Isolation Forest:")
display(df_result["IF_Anomaly"].value_counts().sort_index())

print("\nPersentase hasil Isolation Forest:")
display(
    (df_result["IF_Anomaly"].value_counts(normalize=True).sort_index() * 100)
    .round(2)
    .astype(str) + "%"
)

print("\nTop 10 transaksi paling anomali menurut Isolation Forest:")
display(
    df_result
    .sort_values("IF_Score", ascending=False)
    .head(10)
)
````

### Local Outlier Factor

````python
lof_model = LocalOutlierFactor(
    n_neighbors=20,
    contamination=contamination_rate
)

lof_pred = lof_model.fit_predict(X_model_clean)
lof_score = -lof_model.negative_outlier_factor_

df_result["LOF_Anomaly"] = np.where(lof_pred == -1, 1, 0)
df_result["LOF_Score"] = lof_score

print("Jumlah hasil LOF:")
display(df_result["LOF_Anomaly"].value_counts().sort_index())

print("\nPersentase hasil LOF:")
display(
    (df_result["LOF_Anomaly"].value_counts(normalize=True).sort_index() * 100)
    .round(2)
    .astype(str) + "%"
)

print("\nTop 10 transaksi paling anomali menurut LOF:")
display(
    df_result
    .sort_values("LOF_Score", ascending=False)
    .head(10)
)
````

### One-Class SVM

````python
ocsvm_model = OneClassSVM(
    kernel="rbf",
    gamma="scale",
    nu=contamination_rate
)

ocsvm_model.fit(X_model_clean)

ocsvm_pred = ocsvm_model.predict(X_model_clean)
ocsvm_score = -ocsvm_model.decision_function(X_model_clean)

df_result["OCSVM_Anomaly"] = np.where(ocsvm_pred == -1, 1, 0)
df_result["OCSVM_Score"] = ocsvm_score

print("Jumlah hasil One-Class SVM:")
display(df_result["OCSVM_Anomaly"].value_counts().sort_index())

print("\nPersentase hasil One-Class SVM:")
display(
    (df_result["OCSVM_Anomaly"].value_counts(normalize=True).sort_index() * 100)
    .round(2)
    .astype(str) + "%"
)

print("\nTop 10 transaksi paling anomali menurut One-Class SVM:")
display(
    df_result
    .sort_values("OCSVM_Score", ascending=False)
    .head(10)
)
````

### MCD / Elliptic Envelope

````python
mcd_model = EllipticEnvelope(
    contamination=contamination_rate,
    random_state=random_state
)

mcd_model.fit(X_model_clean)

mcd_pred = mcd_model.predict(X_model_clean)
mcd_score = -mcd_model.decision_function(X_model_clean)

df_result["MCD_Anomaly"] = np.where(mcd_pred == -1, 1, 0)
df_result["MCD_Score"] = mcd_score

print("Jumlah hasil MCD / Elliptic Envelope:")
display(df_result["MCD_Anomaly"].value_counts().sort_index())

print("\nPersentase hasil MCD / Elliptic Envelope:")
display(
    (df_result["MCD_Anomaly"].value_counts(normalize=True).sort_index() * 100)
    .round(2)
    .astype(str) + "%"
)

print("\nTop 10 transaksi paling anomali menurut MCD / Elliptic Envelope:")
display(
    df_result
    .sort_values("MCD_Score", ascending=False)
    .head(10)
)
````

### Model comparison dan ensemble voting

````python
anomaly_columns = [
    "IF_Anomaly",
    "LOF_Anomaly",
    "OCSVM_Anomaly",
    "MCD_Anomaly"
]

df_result["AnomalyVoteCount"] = df_result[anomaly_columns].sum(axis=1)

df_result["Ensemble_Anomaly"] = np.where(
    df_result["AnomalyVoteCount"] >= 2,
    1,
    0
)

model_summary = pd.DataFrame({
    "Model": [
        "Isolation Forest",
        "LOF",
        "One-Class SVM",
        "MCD / Elliptic Envelope",
        "Ensemble Voting"
    ],
    "Anomaly_Count": [
        df_result["IF_Anomaly"].sum(),
        df_result["LOF_Anomaly"].sum(),
        df_result["OCSVM_Anomaly"].sum(),
        df_result["MCD_Anomaly"].sum(),
        df_result["Ensemble_Anomaly"].sum()
    ],
    "Anomaly_Percentage": [
        df_result["IF_Anomaly"].mean() * 100,
        df_result["LOF_Anomaly"].mean() * 100,
        df_result["OCSVM_Anomaly"].mean() * 100,
        df_result["MCD_Anomaly"].mean() * 100,
        df_result["Ensemble_Anomaly"].mean() * 100
    ]
})

model_summary["Anomaly_Percentage"] = model_summary["Anomaly_Percentage"].round(2)

print("Model comparison:")
display(model_summary)

print("\nDistribusi jumlah vote anomaly:")
display(df_result["AnomalyVoteCount"].value_counts().sort_index())

print("\nJumlah hasil Ensemble_Anomaly:")
display(df_result["Ensemble_Anomaly"].value_counts().sort_index())

print("\nTop transaksi berdasarkan jumlah vote anomaly:")
display(
    df_result
    .sort_values(
        ["AnomalyVoteCount", "IF_Score", "LOF_Score", "OCSVM_Score", "MCD_Score"],
        ascending=False
    )
    .head(20)
)
````

AnomalyVoteCount
= berapa banyak model yang menganggap transaksi anomali

### Normalisasi score tiap model

````python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler

if "df_result" not in globals():
    raise ValueError("df_result belum ada. Jalankan modeling dan ensemble voting dulu.")

score_columns = [
    "IF_Score",
    "LOF_Score",
    "OCSVM_Score",
    "MCD_Score"
]

missing_score_columns = [col for col in score_columns if col not in df_result.columns]

if len(missing_score_columns) > 0:
    raise ValueError(f"Score columns belum ada: {missing_score_columns}")

score_scaler = MinMaxScaler()

normalized_scores = score_scaler.fit_transform(df_result[score_columns])

normalized_score_columns = [
    "IF_Score_Norm",
    "LOF_Score_Norm",
    "OCSVM_Score_Norm",
    "MCD_Score_Norm"
]

df_result[normalized_score_columns] = normalized_scores

df_result["Average_AnomalyScore_Norm"] = df_result[normalized_score_columns].mean(axis=1)

print("Contoh hasil normalisasi anomaly score:")
display(
    df_result[
        score_columns + normalized_score_columns + ["Average_AnomalyScore_Norm"]
    ].head(10)
)

print("Statistik normalized anomaly score:")
display(df_result[normalized_score_columns + ["Average_AnomalyScore_Norm"]].describe())
````

**Risk Scoring System**

Setelah proses anomaly detection dilakukan menggunakan beberapa model, tahap berikutnya adalah membuat sistem penilaian risiko atau `RiskScore`. Tahap ini bertujuan untuk mengubah hasil deteksi anomali menjadi skor risiko yang lebih mudah dibaca dan diinterpretasikan.

Pada project ini, dataset tidak memiliki label fraud asli. Oleh karena itu, hasil model tidak digunakan untuk menyatakan bahwa suatu transaksi adalah fraud secara pasti. Hasil model digunakan untuk menandai transaksi yang memiliki pola tidak biasa atau berisiko lebih tinggi dibandingkan transaksi lain.

Model anomaly detection yang digunakan adalah:

```text
Isolation Forest
Local Outlier Factor
One-Class SVM
MCD / Elliptic Envelope
```

Masing-masing model menghasilkan prediksi apakah suatu transaksi dianggap normal atau anomali. Prediksi tersebut kemudian digabungkan menggunakan pendekatan ensemble voting.

**Anomaly Vote Count**

Setiap model memberikan vote anomali dengan nilai:

```text
1 = transaksi dianggap anomali
0 = transaksi dianggap normal
```

Kemudian seluruh vote dari empat model dijumlahkan menjadi fitur:

```text
AnomalyVoteCount
```

Nilai `AnomalyVoteCount` menunjukkan jumlah model yang menganggap suatu transaksi sebagai anomali.

Interpretasinya adalah:

```text
0 vote = tidak ada model yang menganggap transaksi anomali
1 vote = hanya satu model yang menganggap transaksi anomali
2 vote = dua model menganggap transaksi anomali
3 vote = tiga model menganggap transaksi anomali
4 vote = semua model menganggap transaksi anomali
```

Dalam ensemble voting, transaksi ditandai sebagai anomali jika minimal dua model menganggap transaksi tersebut sebagai anomali.

Aturan yang digunakan adalah:

```text
AnomalyVoteCount >= 2  maka Ensemble_Anomaly = 1
AnomalyVoteCount < 2   maka Ensemble_Anomaly = 0
```

Dengan aturan ini, model tidak hanya bergantung pada satu algoritma. Transaksi baru dianggap sebagai anomali ensemble jika terdapat kesepakatan dari minimal dua model.

**Normalisasi Anomaly Score**

Selain menghasilkan label anomali, setiap model juga menghasilkan anomaly score. Score ini menunjukkan tingkat keanehan transaksi menurut masing-masing model.

Namun, score dari setiap model memiliki skala yang berbeda. Sebagai contoh, score dari Isolation Forest, LOF, One-Class SVM, dan MCD tidak berada pada rentang angka yang sama. Oleh karena itu, score dari setiap model perlu dinormalisasi terlebih dahulu.

Normalisasi dilakukan menggunakan MinMaxScaler agar setiap score berada pada rentang:

```text
0 sampai 1
```

Setelah score tiap model dinormalisasi, dibuat rata-rata anomaly score:

```text
Average_AnomalyScore_Norm
```

Fitur ini menunjukkan rata-rata tingkat keanehan transaksi berdasarkan score dari seluruh model setelah dinormalisasi.

**Vote Score**

Selain anomaly score, dibuat juga fitur:

```text
VoteScore
```

Fitur ini dihitung dari jumlah vote anomali dibagi jumlah model.

Rumus konsepnya adalah:

```text
VoteScore = AnomalyVoteCount / 4
```

Karena jumlah model yang digunakan adalah empat, maka nilai `VoteScore` berada pada rentang 0 sampai 1.

Contoh:

```text
AnomalyVoteCount = 0
VoteScore = 0 / 4 = 0

AnomalyVoteCount = 2
VoteScore = 2 / 4 = 0,5

AnomalyVoteCount = 4
VoteScore = 4 / 4 = 1
```

Semakin tinggi nilai `VoteScore`, semakin banyak model yang setuju bahwa transaksi tersebut memiliki pola anomali.

**Perhitungan RiskScore**

RiskScore dibuat dengan menggabungkan dua komponen utama:

```text
VoteScore
Average_AnomalyScore_Norm
```

Bobot yang digunakan adalah:

```text
70% VoteScore
30% Average_AnomalyScore_Norm
```

Rumus konsepnya adalah:

```text
RiskScore = ((0,7 x VoteScore) + (0,3 x Average_AnomalyScore_Norm)) x 100
```

VoteScore diberi bobot lebih besar karena kesepakatan antar model dianggap sebagai sinyal yang lebih kuat. Jika beberapa model berbeda sama-sama menandai transaksi sebagai anomali, maka transaksi tersebut lebih layak diberi skor risiko yang lebih tinggi.

Sementara itu, `Average_AnomalyScore_Norm` tetap digunakan agar RiskScore tidak hanya berdasarkan jumlah vote, tetapi juga mempertimbangkan seberapa kuat transaksi tersebut terlihat anomali berdasarkan score model.

Dengan pendekatan ini, RiskScore berada pada rentang 0 sampai 100.

**RiskLevel**

Setelah RiskScore dihitung, transaksi dikelompokkan ke dalam beberapa level risiko.

Aturan yang digunakan adalah:

```text
RiskScore < 25       = Low
RiskScore 25 - < 50  = Low-Medium
RiskScore 50 - < 75  = Medium
RiskScore >= 75      = High
```

Interpretasi setiap level adalah:

```text
Low
Transaksi memiliki pola yang relatif normal dan tidak banyak model yang menandainya sebagai anomali.

Low-Medium
Transaksi mulai menunjukkan pola yang sedikit berbeda, tetapi tingkat risikonya belum terlalu tinggi.

Medium
Transaksi memiliki pola yang cukup tidak biasa dan perlu diperhatikan lebih lanjut.

High
Transaksi memiliki pola yang sangat tidak biasa dan menjadi prioritas untuk ditinjau.
```

RiskLevel tidak digunakan untuk menyatakan transaksi sebagai fraud secara pasti. RiskLevel hanya digunakan sebagai indikator prioritas risiko dalam konteks anomaly detection.

**Hasil Risk Scoring**

Berdasarkan hasil risk scoring, distribusi RiskLevel adalah:

```text
Low         = 2260 transaksi atau 89,97%
Low-Medium  = 207 transaksi atau 8,24%
Medium      = 27 transaksi atau 1,07%
High        = 18 transaksi atau 0,72%
```

Sebagian besar transaksi berada pada kategori `Low`, yang menunjukkan bahwa mayoritas transaksi memiliki pola relatif normal. Sementara itu, transaksi dengan kategori `Medium` dan `High` jumlahnya jauh lebih sedikit, sehingga dapat menjadi prioritas untuk dianalisis lebih lanjut.

Distribusi RiskLevel berdasarkan Ensemble_Anomaly menunjukkan bahwa:

```text
Seluruh transaksi High termasuk Ensemble_Anomaly
Seluruh transaksi Medium termasuk Ensemble_Anomaly
Sebagian transaksi Low-Medium termasuk Ensemble_Anomaly
Seluruh transaksi Low tidak termasuk Ensemble_Anomaly
```

Hal ini menunjukkan bahwa semakin tinggi RiskLevel, semakin besar kemungkinan transaksi tersebut juga ditandai sebagai anomali oleh ensemble model.

**Hubungan AnomalyVoteCount dan RiskScore**

Hasil rata-rata RiskScore berdasarkan jumlah vote anomali menunjukkan pola yang sesuai:

```text
0 vote = rata-rata RiskScore sekitar 4,79
1 vote = rata-rata RiskScore sekitar 25,23
2 vote = rata-rata RiskScore sekitar 44,49
3 vote = rata-rata RiskScore sekitar 64,35
4 vote = rata-rata RiskScore sekitar 84,20
```

Pola tersebut menunjukkan bahwa semakin banyak model yang memberikan vote anomali, maka nilai RiskScore cenderung semakin tinggi.

Dengan demikian, RiskScore berhasil merepresentasikan tingkat risiko transaksi berdasarkan kombinasi antara jumlah vote anomali dan anomaly score dari masing-masing model.

**Keputusan Akhir**

Pada tahap ini, fitur akhir yang digunakan untuk interpretasi risiko adalah:

```text
AnomalyVoteCount
Ensemble_Anomaly
RiskScore
RiskLevel
```

`AnomalyVoteCount` menunjukkan jumlah model yang menganggap transaksi sebagai anomali.

`Ensemble_Anomaly` menunjukkan hasil keputusan ensemble voting.

`RiskScore` menunjukkan skor risiko transaksi dalam rentang 0 sampai 100.

`RiskLevel` menunjukkan kategori risiko transaksi berdasarkan RiskScore.

**Ringkasan**

Risk Scoring System dibuat untuk mengubah hasil anomaly detection menjadi informasi risiko yang lebih mudah dipahami. Sistem ini menggabungkan hasil voting dari empat model anomaly detection dan anomaly score yang telah dinormalisasi.

RiskScore dihitung menggunakan bobot 70% dari VoteScore dan 30% dari rata-rata normalized anomaly score. Setelah itu, transaksi dikelompokkan menjadi empat kategori risiko, yaitu `Low`, `Low-Medium`, `Medium`, dan `High`.

Dengan pendekatan ini, hasil anomaly detection tidak hanya berupa label normal atau anomali, tetapi juga dilengkapi dengan skor dan level risiko yang lebih informatif untuk kebutuhan analisis dan dashboard.

## Explainability / Interpretability

````python
import numpy as np
import pandas as pd
from IPython.display import display

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

try:
    from xgboost import XGBClassifier
except ImportError:
    raise ImportError("XGBoost belum terinstall. Jalankan dulu: pip install xgboost")

if "X_model_clean" not in globals():
    raise ValueError("X_model_clean belum ada. Jalankan dulu modeling setup.")

if "df_result" not in globals():
    raise ValueError("df_result belum ada. Jalankan dulu anomaly modeling dan risk scoring.")

if "Ensemble_Anomaly" not in df_result.columns:
    raise ValueError("Ensemble_Anomaly belum ada. Jalankan dulu ensemble voting.")

X_surrogate = X_model_clean.copy()
y_surrogate = df_result["Ensemble_Anomaly"].copy()

print("Shape X_surrogate:")
print(X_surrogate.shape)

print("\nDistribusi target Ensemble_Anomaly:")
display(y_surrogate.value_counts().sort_index())

print("\nPersentase target Ensemble_Anomaly:")
display(
    (y_surrogate.value_counts(normalize=True).sort_index() * 100)
    .round(2)
    .astype(str) + "%"
)
````

### Train test split

````python
X_train, X_test, y_train, y_test = train_test_split(
    X_surrogate,
    y_surrogate,
    test_size=0.2,
    random_state=42,
    stratify=y_surrogate
)

print("Shape X_train:")
print(X_train.shape)

print("\nShape X_test:")
print(X_test.shape)

print("\nDistribusi y_train:")
display(y_train.value_counts().sort_index())

print("\nDistribusi y_test:")
display(y_test.value_counts().sort_index())
````

### Train XGBoost surrogate model

````python
negative_count = (y_train == 0).sum()
positive_count = (y_train == 1).sum()

scale_pos_weight = negative_count / positive_count

xgb_surrogate = XGBClassifier(
    n_estimators=300,
    max_depth=3,
    learning_rate=0.05,
    subsample=0.9,
    colsample_bytree=0.9,
    objective="binary:logistic",
    eval_metric="logloss",
    random_state=42,
    scale_pos_weight=scale_pos_weight
)

xgb_surrogate.fit(X_train, y_train)

print("XGBoost surrogate model berhasil dilatih.")
print("\nscale_pos_weight:")
print(scale_pos_weight)
````

### Evaluasi surrogate model

````python
y_pred = xgb_surrogate.predict(X_test)
y_pred_proba = xgb_surrogate.predict_proba(X_test)[:, 1]

evaluation_result = pd.DataFrame({
    "Metric": [
        "Accuracy",
        "Precision",
        "Recall",
        "F1 Score"
    ],
    "Score": [
        accuracy_score(y_test, y_pred),
        precision_score(y_test, y_pred, zero_division=0),
        recall_score(y_test, y_pred, zero_division=0),
        f1_score(y_test, y_pred, zero_division=0)
    ]
})

evaluation_result["Score"] = evaluation_result["Score"].round(4)

print("Evaluasi XGBoost Surrogate Model:")
display(evaluation_result)

print("\nConfusion Matrix:")
display(
    pd.DataFrame(
        confusion_matrix(y_test, y_pred),
        index=["Actual_Normal", "Actual_Anomaly"],
        columns=["Pred_Normal", "Pred_Anomaly"]
    )
)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, zero_division=0))
````

### Tambahkan Top SHAP Feature per transaksi kalau SHAP sudah ada

````python
if "shap_values" in globals() and "X_surrogate" in globals():
    shap_array = shap_values

    if isinstance(shap_array, list):
        if len(shap_array) > 1:
            shap_array = shap_array[1]
        else:
            shap_array = shap_array[0]

    shap_array = np.array(shap_array)

    if shap_array.ndim == 3:
        if shap_array.shape[2] == 2:
            shap_array = shap_array[:, :, 1]
        elif shap_array.shape[0] == 2:
            shap_array = shap_array[1]
        else:
            shap_array = np.squeeze(shap_array)

    if shap_array.shape[0] != df_final_result.shape[0]:
        raise ValueError("Jumlah baris SHAP values tidak sama dengan df_final_result.")

    top_k = 3
    top_shap_indices = np.argsort(np.abs(shap_array), axis=1)[:, ::-1][:, :top_k]

    for i in range(top_k):
        df_final_result[f"Top_SHAP_Feature_{i+1}"] = [
            X_surrogate.columns[idx]
            for idx in top_shap_indices[:, i]
        ]

        df_final_result[f"Top_SHAP_Value_{i+1}"] = [
            shap_array[row_idx, feature_idx]
            for row_idx, feature_idx in enumerate(top_shap_indices[:, i])
        ]

        df_final_result[f"Top_SHAP_Value_{i+1}"] = df_final_result[f"Top_SHAP_Value_{i+1}"].round(4)

    print("Top SHAP features berhasil ditambahkan ke df_final_result.")
    print("Catatan: SHAP value positif mendorong prediksi ke arah anomaly, SHAP value negatif mendorong ke arah normal.")

    display(
        df_final_result
        .sort_values("RiskScore", ascending=False)
        [
            [
                "TransactionID",
                "RiskScore",
                "RiskLevel",
                "AnomalyVoteCount",
                "Top_SHAP_Feature_1",
                "Top_SHAP_Value_1",
                "Top_SHAP_Feature_2",
                "Top_SHAP_Value_2",
                "Top_SHAP_Feature_3",
                "Top_SHAP_Value_3"
            ]
        ]
        .head(20)
    )
else:
    print("SHAP belum tersedia. Cell ini dilewati.")
````
